# Day 09 — The Process Lifecycle, Visualized

A process moves through states as the OS admits, dispatches, preempts, blocks, and reaps
it. Every `dispatch` is a context switch. Let's drive one through a realistic run.

```python
TRANSITIONS = {
    ('new', 'admit'): 'ready',
    ('ready', 'dispatch'): 'running',
    ('running', 'timeout'): 'ready',
    ('running', 'wait'): 'waiting',
    ('waiting', 'wakeup'): 'ready',
    ('running', 'exit'): 'terminated',
}

state = 'new'; history = [state]; switches = 0
events = ['admit', 'dispatch', 'timeout', 'dispatch', 'wait', 'wakeup', 'dispatch', 'exit']
for e in events:
    state = TRANSITIONS[(state, e)]; history.append(state)
    if e == 'dispatch': switches += 1
    print(f'{e:9} -> {state}')
print('\ncontext switches (CPU grants):', switches)
```

## The run as a timeline of states

```python
import matplotlib.pyplot as plt

order = ['terminated', 'waiting', 'ready', 'running', 'new']
ys = [order.index(s) for s in history]
plt.figure(figsize=(8, 3))
plt.step(range(len(history)), ys, where='post', marker='o')
plt.yticks(range(len(order)), order)
plt.xlabel('event #'); plt.title('One process walking its lifecycle')
plt.grid(alpha=0.3); plt.show()
```

## Takeaways

- A process is a state machine; the OS only allows legal transitions.
- A blocked (`waiting`) process must be woken to `ready` before it can run again.
- Each `dispatch` is a context switch — cheap, but not free.
