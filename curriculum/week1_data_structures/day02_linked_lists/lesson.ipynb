# Day 02 — Linked Lists

A dynamic array stores items shoulder-to-shoulder in one block, so inserting in the
middle shoves everything after it over. A **linked list** threads items with pointers
instead: each node points to its neighbours. Insert/delete at a known node is O(1).

## A node is just a value plus two pointers

`prev  ⟵  [ value ]  ⟶  next`. Chain them and you get a list you can walk both ways.

```python
class Node:
    def __init__(self, value):
        self.value = value; self.prev = None; self.next = None

def build(values):
    head = None; tail = None
    for v in values:
        n = Node(v)
        if tail: tail.next = n; n.prev = tail
        else: head = n
        tail = n
    return head

def forward(head):
    out = []
    while head: out.append(head.value); head = head.next
    return out

head = build([1, 2, 3, 4])
print('forward:', forward(head))
```

## Draw the chain

```python
import matplotlib.pyplot as plt

def draw(head, title):
    vals = forward(head)
    fig, ax = plt.subplots(figsize=(1.6 * len(vals) + 1, 1.8))
    for i, v in enumerate(vals):
        ax.add_patch(plt.Rectangle((i * 2, 0), 1.2, 1, fill=True,
                                   facecolor='#4C72B0', edgecolor='black'))
        ax.text(i * 2 + 0.6, 0.5, str(v), ha='center', va='center', color='white')
        if i < len(vals) - 1:
            ax.annotate('', xy=(i * 2 + 2, 0.65), xytext=(i * 2 + 1.2, 0.65),
                        arrowprops=dict(arrowstyle='->'))
            ax.annotate('', xy=(i * 2 + 1.2, 0.35), xytext=(i * 2 + 2, 0.35),
                        arrowprops=dict(arrowstyle='->', color='gray'))
    ax.set_xlim(-0.5, 2 * len(vals)); ax.set_ylim(-0.3, 1.3)
    ax.axis('off'); ax.set_title(title); plt.show()

draw(head, 'blue = next,  gray = prev')
```

## Insert 99 after the 2nd node: two pointers change, nothing shifts

Compare that to a Python list, where `lst.insert(2, 99)` slides every later element.

```python
def insert_after_index(head, index, value):
    cur = head
    for _ in range(index): cur = cur.next
    n = Node(value)
    n.prev = cur; n.next = cur.next
    if cur.next: cur.next.prev = n
    cur.next = n
    return head        # only cur.next, n.prev/next, and one neighbour touched

head = insert_after_index(head, 1, 99)
print('after insert:', forward(head))
draw(head, 'inserted 99 after index 1 — O(1) rewire')
```

## Takeaways

- Linked list: O(1) insert/delete at a known node, O(n) to *find* the i-th item.
- Dynamic array: O(1) random access, O(n) middle insert.
- *Doubly* linked (both `prev` and `next`) is what makes O(1) deletion and backward
  walking possible.
