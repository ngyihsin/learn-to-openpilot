# Day 07 — Graphs, BFS & Grid Pathfinding, Visualized

BFS explores in rings of increasing distance, so on an unweighted graph it finds the
shortest path. The same idea routes around walls on a grid — a toy motion planner.

## BFS distances from a start node

```python
from collections import deque

adj = {0: [1, 2], 1: [0, 3], 2: [0, 3], 3: [1, 2, 4], 4: [3]}

def bfs_dist(adj, start):
    dist = {start: 0}; q = deque([start])
    while q:
        u = q.popleft()
        for v in adj[u]:
            if v not in dist: dist[v] = dist[u] + 1; q.append(v)
    return dist

print('distance from 0:', bfs_dist(adj, 0))
```

## Shortest path on a grid ('#' = wall)

```python
import matplotlib.pyplot as plt
import numpy as np

grid = ['.....', '.###.', '.....']
start, goal = (0, 0), (2, 4)

def grid_path(grid, start, goal):
    R, C = len(grid), len(grid[0])
    ok = lambda r, c: 0 <= r < R and 0 <= c < C and grid[r][c] != '#'
    q = deque([start]); parent = {start: None}
    while q:
        r, c = q.popleft()
        if (r, c) == goal: break
        for dr, dc in ((-1, 0), (1, 0), (0, -1), (0, 1)):
            n = (r + dr, c + dc)
            if ok(*n) and n not in parent: parent[n] = (r, c); q.append(n)
    if goal not in parent: return None
    path = [goal]
    while parent[path[-1]] is not None: path.append(parent[path[-1]])
    return path[::-1]

path = grid_path(grid, start, goal)
img = np.array([[0 if ch == '#' else 1 for ch in row] for row in grid], dtype=float)
for (r, c) in path: img[r][c] = 0.5
plt.figure(figsize=(5, 3))
plt.imshow(img, cmap='Blues'); plt.title(f'shortest path: {len(path)} cells')
plt.xticks([]); plt.yticks([]); plt.show()
print('path:', path)
```

## Takeaways

- BFS (queue) = shortest path by edge count; DFS (stack/recursion) = deep exploration.
- A grid is just a graph with up/down/left/right edges — the same BFS routes around walls.
- Add edge weights and BFS becomes Dijkstra, with the priority queue from Day 06.
