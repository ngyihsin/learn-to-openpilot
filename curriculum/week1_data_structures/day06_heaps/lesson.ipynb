# Day 06 — Heaps, Visualized

A binary heap is a complete tree packed into an array: children of index `i` are `2i+1`
and `2i+2`, and every parent ≤ its children. The min is always at index 0.

```python
def sift_up(a, i):
    while i > 0 and a[i] < a[(i - 1) // 2]:
        a[i], a[(i - 1) // 2] = a[(i - 1) // 2], a[i]; i = (i - 1) // 2

heap = []
for v in [5, 3, 8, 1, 9, 2, 7]:
    heap.append(v); sift_up(heap, len(heap) - 1)
print('heap array:', heap, ' -> min at index 0:', heap[0])
```

## The array *is* a tree

```python
import matplotlib.pyplot as plt

def draw_heap(a, title):
    fig, ax = plt.subplots(figsize=(7, 4))
    pos = {}
    import math
    for i in range(len(a)):
        depth = int(math.log2(i + 1)); span = 2 ** depth
        offset = (i + 1) - span
        x = (offset + 0.5) / span; pos[i] = (x, -depth)
    for i in range(len(a)):
        for c in (2 * i + 1, 2 * i + 2):
            if c < len(a): ax.plot([pos[i][0], pos[c][0]], [pos[i][1], pos[c][1]], color='gray', zorder=1)
    for i, (x, y) in pos.items():
        ax.scatter([x], [y], s=650, color='#55A868', zorder=2)
        ax.text(x, y, str(a[i]), ha='center', va='center', color='white', zorder=3)
    ax.set_title(title); ax.axis('off'); plt.show()

draw_heap(heap, 'Min-heap: every parent <= its children')
```

## Heapsort: pop the min until empty

```python
import heapq
data = [5, 3, 8, 1, 9, 2, 7, 4, 6, 0]
h = data[:]; heapq.heapify(h)
ordered = [heapq.heappop(h) for _ in range(len(h))]
print('heapsorted:', ordered)
assert ordered == sorted(data)
```

## Takeaways

- O(1) peek-min, O(log n) push/pop, O(n) to heapify — all with pure index math, no pointers.
- Pop-until-empty is heapsort: O(n log n), in place.
- A heap is the priority queue behind Dijkstra/A* and the SJF scheduler (Day 08).
