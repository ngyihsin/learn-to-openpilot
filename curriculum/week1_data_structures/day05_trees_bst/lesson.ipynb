# Day 05 — Binary Search Trees, Visualized

Smaller left, larger right. That invariant makes search O(height) and an in-order walk
come out **sorted** — unless you insert already-sorted data and the tree collapses.

```python
class N:
    def __init__(self, v): self.v = v; self.l = None; self.r = None

def insert(root, v):
    if root is None: return N(v)
    if v < root.v: root.l = insert(root.l, v)
    elif v > root.v: root.r = insert(root.r, v)
    return root

def in_order(root, out):
    if root: in_order(root.l, out); out.append(root.v); in_order(root.r, out)

root = None
for v in [5, 3, 8, 1, 4, 7, 9]: root = insert(root, v)
out = []; in_order(root, out); print('in-order traversal (sorted!):', out)
```

## Draw it

```python
import matplotlib.pyplot as plt

def draw(root, title):
    pos = {}; counter = [0]
    def assign(node, depth):
        if node is None: return
        assign(node.l, depth + 1)
        pos[node] = (counter[0], -depth); counter[0] += 1
        assign(node.r, depth + 1)
    assign(root, 0)
    fig, ax = plt.subplots(figsize=(7, 4))
    def edges(node):
        if node is None: return
        for c in (node.l, node.r):
            if c: ax.plot([pos[node][0], pos[c][0]], [pos[node][1], pos[c][1]], color='gray', zorder=1)
        edges(node.l); edges(node.r)
    edges(root)
    for node, (x, y) in pos.items():
        ax.scatter([x], [y], s=650, color='#4C72B0', zorder=2)
        ax.text(x, y, str(node.v), ha='center', va='center', color='white', zorder=3)
    ax.set_title(title); ax.axis('off'); plt.show()

draw(root, 'A reasonably balanced BST')
```

## The failure mode: insert already-sorted data

The tree degenerates into a linked list — height n−1, and search is back to O(n).

```python
bad = None
for v in [1, 2, 3, 4, 5]: bad = insert(bad, v)
draw(bad, 'Sorted insert -> degenerate chain (this is why balanced trees exist)')
```

## Takeaways

- In-order traversal of a BST is sorted; search/insert/delete are O(height).
- Balanced height is O(log n); a degenerate tree is O(n). Real systems use *self-balancing*
  trees (red-black, B-trees) to guarantee the good case.
