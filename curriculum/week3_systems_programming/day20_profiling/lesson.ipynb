# Day 20 — Profiling: Find the O(n^2), Visualized

Correct but slow is still slow. The fix is almost always deleting an accidental nested
loop. Measure it, then watch the linear version leave the quadratic one in the dust.

```python
import timeit, random

def naive_dup(items):
    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if items[i] == items[j]: return True
    return False

def fast_dup(items):
    seen = set()
    for x in items:
        if x in seen: return True
        seen.add(x)
    return False

data = list(range(2000))            # all distinct -> worst case
tn = timeit.timeit(lambda: naive_dup(data), number=3) / 3
tf = timeit.timeit(lambda: fast_dup(data), number=3) / 3
print(f'naive O(n^2): {tn*1000:8.2f} ms')
print(f'fast  O(n):   {tf*1000:8.2f} ms   ({tn/tf:.0f}x faster)')
```

## Runtime vs input size: the curves diverge

```python
import matplotlib.pyplot as plt

sizes = [250, 500, 1000, 2000, 4000]
naive_t, fast_t = [], []
for n in sizes:
    d = list(range(n))
    naive_t.append(timeit.timeit(lambda: naive_dup(d), number=2) / 2)
    fast_t.append(timeit.timeit(lambda: fast_dup(d), number=20) / 20)
plt.figure(figsize=(7, 4))
plt.plot(sizes, naive_t, marker='o', label='naive O(n^2)')
plt.plot(sizes, fast_t, marker='s', label='fast O(n)')
plt.xlabel('input size'); plt.ylabel('seconds'); plt.yscale('log')
plt.title('Same answer, wildly different scaling'); plt.legend(); plt.grid(alpha=0.3); plt.show()
```

## Takeaways

- Profile before optimizing; the hot spot is rarely where you'd guess.
- `x in set` is O(1) vs `x in list` O(n) — swapping the container often IS the fix.
- Big-O predicts how badly it hurts as data grows.
